```python
#making the data numeric by one hot encoding
import pandas as pd
df_new = pd.get_dummies(df,columns=['Sex','ChestPainType','RestingECG','ExerciseAngina','ST_Slope'])    
df_new
```

```python
df_new.columns
```

```python
#splitting data into X(all features)and y(target variable)
X = df_new[['Age', 'RestingBP', 'Cholesterol', 'FastingBS', 'MaxHR', 'Oldpeak',
         'Sex_F', 'Sex_M', 'ChestPainType_ASY',
       'ChestPainType_ATA', 'ChestPainType_NAP', 'ChestPainType_TA',
       'RestingECG_LVH', 'RestingECG_Normal', 'RestingECG_ST',
       'ExerciseAngina_N', 'ExerciseAngina_Y', 'ST_Slope_Down',
       'ST_Slope_Flat', 'ST_Slope_Up']]
y = df_new[['HeartDisease']]
```

```python
#train-test split
from sklearn.model_selection import train_test_split
# train_test_split splits the dataset into training and testing datasets. Here I have specified the training size as 70%.
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#model training
from sklearn.linear_model import LogisticRegression
# Creating the logistic regression model, and fitting/training it using the training dataset -
log_reg_model = LogisticRegression(solver='liblinear').fit(X_train, y_train.values.ravel())
```

```python
#predictions
pred = log_reg_model.predict(X_test)
print(pred)
```

```python
#classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, pred))
```

```python
# printing score of the model -
score = log_reg_model.score(X_test, y_test)
print(score)
```
